```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Data from spreadsheet
quarters = ['Q1 21', 'Q2 21', 'Q3 21', 'Q4 21', 'Q1 22', 'Q2 22', 'Q3 22', 'Q4 22']

# ARR Growth
ending_arr = [3777, 5460, 6089, 7278, 8164, 9798, 11496, 13841]

# Customer Growth
eop_customers = [23, 32, 37, 42, 50, 57, 69, 78]

# Revenue vs. OPEX
revenue = [707, 740, 855, 1096, 1212, 1405, 1564, 1769]
opex = [2722, 3476, 4435, 4732, 6259, 7076, 6946, 7562]

# Net Burn Rate
burn_rate = [2224, 3178, 4432, 3643, 5946, 6545, 6346, 6245]
```

```python

# Set seaborn style
sns.set_style("whitegrid")

# Plot ARR Growth
plt.figure(figsize=(10, 5))
plt.plot(quarters, ending_arr, marker='o', linestyle='-', color='b', label="Ending ARR ($000s)")
plt.title("ARR Growth Over Time")
plt.xlabel("Quarter")
plt.ylabel("ARR ($000s)")
plt.xticks(rotation=45)
plt.legend()
plt.show()

# Plot Customer Growth
plt.figure(figsize=(10, 5))
sns.barplot(x=quarters, y=eop_customers, palette="Blues_r")
plt.title("Customer Growth Over Time")
plt.xlabel("Quarter")
plt.ylabel("Number of Customers")
plt.xticks(rotation=45)
plt.show()

# Plot Revenue vs. OPEX
plt.figure(figsize=(10, 5))
x = np.arange(len(quarters))
width = 0.4
plt.bar(x - width/2, revenue, width, label="Revenue ($000s)", color="green")
plt.bar(x + width/2, opex, width, label="OPEX ($000s)", color="red")
plt.title("Revenue vs. Operating Expenses")
plt.xlabel("Quarter")
plt.ylabel("Amount ($000s)")
plt.xticks(ticks=x, labels=quarters, rotation=45)
plt.legend()
plt.show()

# Plot Net Burn Rate
plt.figure(figsize=(10, 5))
plt.plot(quarters, burn_rate, marker='s', linestyle='-', color='r', label="Net Burn Rate ($000s)")
plt.title("Net Burn Rate Over Time")
plt.xlabel("Quarter")
plt.ylabel("Burn Rate ($000s)")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
